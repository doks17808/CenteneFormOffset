```python
import xml.etree.ElementTree as ET

import pandas as pd
```

# XML to pd.DataFrame

###### Removed unnecessary information that is saved from LabelImg. Only need annotations and objects.

# Blank Form

```python
def xmltodf(path):
    xtree = ET.parse(path)
    xroot=xtree.getroot()
    df_cols = ['name', 'xmin','ymin','xmax','ymax']
    rows = []

    for node in xroot:
        x_name = node[0].text
        xmin = node[4][0].text
        ymin = node[4][1].text
        xmax = node[4][2].text
        ymax = node[4][3].text
        rows.append({'name':x_name, 'xmin':xmin, 'ymin':ymin, 'xmax':xmax, 'ymax':ymax})
    df = pd.DataFrame(rows,columns=df_cols)
    df.to_csv(f'{path[:-4]}.csv', index=False)
    return df
```
